```python
import numpy as np
import torch

import pandas as pd
```

```python
def model(W, x, b):
    return x @ W + b
```

```python
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
df = pd.read_csv("../data/boston.csv")
df.head()
```

```python
X = df.drop("TAX", axis=1).to_numpy(dtype='float32')
y = df["TAX"].to_numpy(dtype='float32')
```

```python
X_mean = X.mean(axis=0)
X_std = X.std(axis=0)
X = (X - X_mean) / X_std

y_mean = y.mean()
y_std = y.std()
y = (y - y_mean) / y_std
```

```python
X.shape, y.shape
```

```python
X = torch.from_numpy(X)
y = torch.from_numpy(y.reshape(-1, 1))
```

```python
X.shape
```

```python
w = torch.rand(X.shape[1], 1)
b = torch.rand(1, )
```

```python
w = torch.tensor(w, requires_grad=True)
b = torch.tensor(b, requires_grad=True)
```

```python
for i in range(1500):
    y_hat = model(w, X, b)
    loss = mse(y, y_hat)
    loss.backward()

    with torch.no_grad():
        w -= w.grad * 1e-2
        b -= b.grad * 1e-2
        w.grad.zero_()
        b.grad.zero_()
```
