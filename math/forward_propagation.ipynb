```python
import numpy as np
```

```python
parameters = {}
layers = [3, 4, 2]
activations = ["relu", "sigmoid"]
```

```python
def initialize_layer_parameters(n_in, n_out):
    W = np.random.randn(n_in, n_out) * 0.01
    b = np.zeros((1, n_out))
    return W, b
```

```python
for i in range(1, len(layers)):
    W, b = initialize_layer_parameters(layers[i-1], layers[i])
    parameters[i] = {
        "W": W,
        "b": b
    }
```

```python
parameters
```

```python
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

def relu(Z):
    return np.maximum(Z, 0)
```

```python
def forward_propagation_layer(X, parameters, activations):
    A = X
    layers_size = len(parameters)

    for l in range(1, layers_size+1):
        params = parameters[l]
        W = params["W"]
        b = params["b"]

        Z = A @ W + b
        parameters[l]["Z"] = Z
        
        activation = activations[l-1]
        if activation == "sigmoid":
            A = sigmoid(Z)
        elif activation == "relu":
            A = relu(Z)
        
        parameters[l]["A"] = A
    
    return A
```

```python
X_exemplo = np.random.rand(10, 3) 
A_layer = forward_propagation_layer(X_exemplo, parameters, activations)
print("Output shape:", A_layer.shape)
print(f"\nA:\n{A_layer}")
```
